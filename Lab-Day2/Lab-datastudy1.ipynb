#### Concatenate the three dataframes

```python
column_names=file1.columns
column_names
```

```python
data=pd.DataFrame(columns=column_names)
```

```python
data=pd.concat([file1,file2,file3],axis=0)
```

```python
#reset the index!! 
data=data.reset_index(drop=True)
data
```

#### Data cleaning operations

```python
#Delete the column education and the number of open complaints from the dataframe.
data = data.drop(['education','number of open complaints'], axis=1)
data.head()
```

```python
#Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
data['customer lifetime value'] =  pd.to_numeric(data['customer lifetime value'], errors='coerce')*100
data['customer lifetime value']
```

```python
#Check for duplicate rows in the data and remove if any.
duplicate = data[data.duplicated()]
duplicate
```

```python
data_without_duplicates = data.drop_duplicates()
data_without_duplicates.shape
```

```python
#Filter out the data for customers who have an income of 0 or less.
filtered_data = data_without_duplicates[data_without_duplicates['income']<= 0]
filtered_data
```
